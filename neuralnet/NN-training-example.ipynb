```python
import numpy as np
import csv
from NNengine import Run_NN, Multilayers
```

# Training Neural Network
### Simple example of single-layer NN training on singular data set. To optimize, train for several different data sets/HH simulations.

## Import simulation training values

```python
fn = open('../sim_data/gt_1a.csv', 'r')
f = csv.DictReader(fn)
```

```python
stim = []
voltage = []
for col in f:
    stim.append(float(col['stim']))
    voltage.append(float(col['voltage']))
stim = np.asarray(stim)
voltage = np.asarray(voltage)

#isolate values where stimulus present
start,stop = (np.where(stim!=0)[0][0], np.where(stim!=0)[0][-1])
test_stim = stim[start:stop]
test_voltage = voltage[start:stop]
```

# NN for a square wave
#### We rescale the stimulus using our activation function for use in comparison within NN. Voltage data is normalized for NN accuracy.

```python
raw_stim = stim[start:stop]
test_stim = [1/(1+np.exp(s)) for s in test_stim]
test_voltage = (test_voltage - min(test_voltage))/(max(test_voltage) - min(test_voltage))
```

```python
nouts = [300, len(test_stim)] #single layer neural network with 300 nodes (there are 199 test_voltage values)
nin = len(test_voltage) #number of input nodes/values
```

### Initiate NN object and run

```python
NN = Multilayers(nin, nouts) #initiate our NN class object
stim_pred,loss = Run_NN(NN, test_stim, test_voltage, iter_lim=50, error_thresh=.001)
```

```python
stim_pred
```

#### If we go back to the original stimulus scale, we can see that the output is in the ballpark of our original stimulus: 1.9. To improve accuracy, the error threshold and iteration limits would have to be adjusted. To test this, increase iter_lim and reduce error_thresh in Run_NN().

```python
rescaled = np.log((1 / (stim_pred + 1e-8)) - 1) #should give you input stimulus at original scale
rescaled
```

## Plot the Input Wave vs. Neural Network Prediction
#### We purposely omit values where the system is at rest for efficiency; plots representing the square wave will be a line y=amplitude.

```python
import matplotlib.pyplot as plt
```

```python
fig, (ax0, ax1) = plt.subplots(2, 1, layout='constrained', sharey=True)
#exact time is arbitary; concerned with time steps rather
t = np.arange(1, len(test_stim)+1, 1)
ax0.title.set_text('NN output')
ax0.plot(t, rescaled)
ax0.plot(t, raw_stim, color='red', alpha=.2, label='real input')
ax1.title.set_text('Real input')
ax1.plot(t, raw_stim, color='red')
ax1.set_xlabel('N Time Stamp')
ax0.set_ylabel('Pulse Amplitude')
ax1.set_ylabel('Pulse Amplitude')
ax0.legend()

plt.show()
```

#### The Neural Network returns values that fluctuate around the original input/pulse amplitude value(s). We can use our waveform code to recognize the data as a particular wave type and transform it accordingly to get our final prediction. In the simple square wave case shown here, after matching the data to a square waveform, this transformation simply consists of replacing every data point with the output array's average value.

```python
rescaled_stim = rescaled
rescaled_stim = np.asarray([np.mean(rescaled_stim)]*len(rescaled_stim))
fig2, (ax2, ax3) = plt.subplots(2, 1, layout='constrained', sharey=True)
#exact time is arbitary; concerned with time steps rather
t = np.arange(1, len(test_stim)+1, 1)
ax2.title.set_text('NN output')
ax2.plot(t, rescaled_stim)
ax3.title.set_text('Real input')
ax3.plot(t, raw_stim, color='red')
ax3.set_ylim([raw_stim[0]-.2, raw_stim[0]+.2])
ax3.set_xlabel('Time Stamp N')
ax2.set_ylabel('Pulse Amplitude')
ax3.set_ylabel('Pulse Amplitude')

plt.show()
```

## Fit Output Data to Wave Type using Waveform Code

#### We would generally not know the output of the NN was supposed to be a square waveform, so we would use the following code to identify the output as such before doing the transformation in the previous step. 

```python
from WaveformGenerator import Waveform
```

### Iterate through, fit to different wave types to return best fit:

```python
from scipy.optimize import leastsq
```

```python
'''
Waveform bound methods take the following arguments; see WaveformGenerator.py:
square_wave(frequency, amplitude, duty_cycle) 
sine_wave(frequency, amplitude)
triangular_wave(frequency, amplitude)
gaussian_pulse(amplitude, fwhm_seconds)
'''

errors = {}
wave_objects = {}
#do not need to optimize parameters for square wave since amplitude follows directly from mean of NN output data
wave = Waveform(1, sampling_rate=len(test_stim))
wave.square_wave(1, np.mean(rescaled), duty_cycle=1)
wave_stim = wave.voltage_data
errors['square_wave'] = np.square(np.subtract(wave_stim, rescaled)).mean()
wave_objects['square_wave'] = wave

#Use the following to iterate over other waveforms
N = len(rescaled) # number of data points
t = np.linspace(1, len(rescaled), N) #arbitrary time values
forms = ['sine_wave', 'triangular_wave', 'gaussian_pulse']
init_args = [(1,1), (1,1), (1,len(rescaled)/2)] #for use by optimization

def return_wave(wave_obj, wave_form_name, p1, p2):
    '''
    Runs Waveform bound method on Waveform object and returns voltage data. 
    '''
    wave_func = getattr(wave_obj, wave_form_name)
    wave_func(p1, p2)
    return wave_obj.voltage_data
    
#iterate through different waveforms' outputs, get MSE relative to NN output, load value to dic.
for wave_form, params in zip(forms, init_args):
    wave = Waveform(1, sampling_rate=len(test_stim))
    optimize_func = lambda x: return_wave(wave, wave_form, x[0], x[1]) - rescaled
    est_p1, est_p2 = leastsq(optimize_func, list(params))[0]
    wave_func(est_p1, est_p2)
    wave_stim = wave.voltage_data
    errors[wave_form] = np.square(np.subtract(wave_stim, rescaled)).mean()
    wave_objects[wave_form] = wave

#obtain lowest error to get best fit waveform
best_fit = min(errors, key=errors.get)
print('Best fit waveform is: ' + best_fit + '!')
```

### Plot the waveform fit Against NN output data:

```python
best_fit_data = wave_objects[best_fit].voltage_data
fig = plt.figure()
plt.plot(t, rescaled, color='black', alpha=.5) #NN output
plt.plot(t, best_fit_data, color='red')
plt.ylabel('Wave Amplitude')
plt.xlabel('Time Stamp N')
```

```python
print('MSE for this fit is: ' + str(errors[best_fit]))
```
